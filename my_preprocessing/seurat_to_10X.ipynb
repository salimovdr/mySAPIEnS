```r
library(Seurat)
library(Matrix)
```

```r
obj <- readRDS("~/Datasets/GSE166547/input/GSE166547.rds")
```

```r
DefaultAssay(obj) <- "ATAC"
```

```r
mat <- GetAssayData(obj, assay = "ATAC", slot = "counts")
```

```r
barcodes <- colnames(mat)
write.table(barcodes, "~/Datasets/atac_small/input/barcodes.tsv",
            sep = "\t", row.names = FALSE, col.names = FALSE, quote = FALSE)
```

```r
peak_ids <- rownames(mat)

# Разбиваем "chr1:10109-10357" на chr, start, end
# Сначала делим по ":" и "-"
parts <- do.call(rbind, strsplit(peak_ids, "[:-]"))

peaks_df <- data.frame(
  chr   = parts[, 1],
  start = parts[, 2],
  end   = parts[, 3],
  stringsAsFactors = FALSE
)

# Сохраняем в BED без заголовков
write.table(
  peaks_df,
  file = "~/Datasets/atac_small/input/peaks.bed",
  sep = "\t",
  row.names = FALSE,
  col.names = FALSE,
  quote = FALSE
)
```

```r
writeMM(mat, "~/Datasets/atac_small/input/matrix.mtx")
```

```r
# barcodes — уже получены из colnames(mat)
# mat <- GetAssayData(obj, assay = "ATAC", slot = "counts")

celltype_col <- "celltype"   # ← подставь своё имя столбца

labels_df <- data.frame(
  barcode = barcodes,
  label   = obj@meta.data[barcodes, celltype_col],
  stringsAsFactors = FALSE
)

write.table(
  labels_df,
  file = "~/Datasets/PBMC5k/input/labels.tsv",
  sep = "\t",
  row.names = FALSE,
  col.names = TRUE,
  quote = FALSE
)

```
